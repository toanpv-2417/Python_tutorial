```python
df = pd.read_csv('netflix_titles.csv')
```

```python
df = df.dropna(subset=['country'])
```

```python
df = df.head(5)
```

```python
df
```

```python
df['country'] = [countries[0] for countries in df['country'].str.split(',')]
```

```python
type(df['country'])
```

```python
country_counts = pd.DataFrame(
  df['country'].value_counts().rename_axis(
    'countries').reset_index(name='counts')
  ).sort_values(by=['countries'])
```
