```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.available
plt.style.use('seaborn-whitegrid')
```

```python
cities = pd.read_csv('california_cities.csv')
cities.head()

cities['area_total_km2'].sort_values(ascending=False)
```

```python
#Extract Latd
lat,lon = cities['latd'],cities['longd']
population ,area = cities['population_total'],cities['area_total_km2'],
#Plot using Pylot API

plt.figure(figsize=(8,6))
plt.style.use('seaborn')
plt.scatter(lon,lat,
            c=np.log10(population),cmap='viridis',
            s=area,
            
           )
plt.axis('equal')
plt.xlabel('longtitude')
plt.ylabel('latitude')
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3,7)

#create a lengend for cities size

area_range = [50,100,300,500]

for area in area_range:
    plt.scatter([],[],s=area,
                c= "k",
                alpha=0.4,
               label=str(area) + ' km2'
               )
plt.legend(scatterpoints=1,labelspacing=1.5,title="City area")
plt.title('California ciies')
```
